```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('popular_board_game.csv')
data
```

```python
solo = data[data['min_players'] == 1].reset_index(drop=True)
no_solo = data[data['min_players'] > 1].reset_index(drop=True)
```

```python
# Info needed: number of games in each category, game w/ highest num of ratings, average rating of games rounded to
# 2 decimal places, average playtime rounded to 2 dec places
# Also include gt username below bottom left of table, and save screenshot as table.png
ngames_solo = len(solo)
ngames_nosolo = len(no_solo)
highest_solo = solo.sort_values('num_ratings',ascending=False).loc[0,'name']
highest_nosolo = no_solo.sort_values('num_ratings',ascending=False).loc[0,'name']
avgrate_solo = round(np.mean(solo['avg_rating']),2)
avgrate_nosolo = round(np.mean(no_solo['avg_rating']),2)
avgtime_solo = round(np.mean(solo['playtime_num']),2)
avgtime_nosolo = round(np.mean(no_solo['playtime_num']),2)

solo_all = [ngames_solo,highest_solo,avgrate_solo,avgtime_solo]
nosolo_all = [ngames_nosolo,highest_nosolo,avgrate_nosolo,avgtime_nosolo]
```

```python
solo_all,nosolo_all
```

```python
df_data = pd.DataFrame({'Solo Supported':solo_all, 'Solo Not Supported':nosolo_all}).T
df_data.columns = ['Number of Games','Game with Most Ratings','Average Game Rating','Average Playtime']
df_data
```
